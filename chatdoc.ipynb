```python
#定义chatboc
from langchain_classic.document_loaders import Docx2txtLoader  #docx文档加载器
from langchain_classic.document_loaders import UnstructuredExcelLoader  #xlsx文档加载器
from langchain_classic.text_splitter import CharacterTextSplitter  #文档切割器

from langchain_openai.chat_models import ChatOpenAI
from langchain_classic.retrievers.multi_query import MultiQueryRetriever#多重查询检索模块
from langchain_classic.prompts import ChatPromptTemplate  #prompt相关模块


class ChatDoc():
    def __init__(self):
        self.doc = None
        self.splitText = []
        #为满足对话要求，进行额外定义
        self.template = [
            ("system","你是一个文档处理助手，你从不说自己是一个大模型或者AI助手，你能够根据上下文内容来继续回答问题。\n 上下文内容：\n{context}\n"),
            ("human","你好！"),
            ("ai","你好！"),
            ("human","{question}"),
        ]
        self.prompt = ChatPromptTemplate.from_messages(self.template) 

    def getFile(self):
        doc = self.doc
        loaders = {
            "docx": Docx2txtLoader,
            "xlsx": UnstructuredExcelLoader,
        }
        file_extension = doc.split(".")[-1]
        loader_class = loaders.get(file_extension)
        if loader_class:
            try:
                #读取文件
                loader = loader_class(doc)
                text = loader.load()
                return text
            except Exception as e:
                print(f"Error loading{file_extension} files: {e}")
        else:
            print(f"Unsupported file extension: {file_extension}")
    def splitsentence(self):
        full_text = self.getFile()
        if full_text != None:
            #对文档进行切割
            splitter = CharacterTextSplitter(
                #seperator = "\n\n", #默认为双换行符
                chunk_size = 150, #每个文本块的最大长度
                chunk_overlap = 30 #相邻块的重合长度
            )
            texts = splitter.split_documents(full_text) #
            self.splitText = texts
    #提问并找到相关的文本块
    def askAndFindFiles(self,question):
        db = Chroma.from_documents(docs, embeddings)
        llm = ChatOpenAI(                            #chat模型调用地址进行修改，与embedding模型地址不同
            temperature=0,
            model="Pro/deepseek-ai/DeepSeek-V3.2",
            api_key="sk-oizcnnagplmevyplnagoxhggqqmpfprfhfoupaqvfvwonxes",
            base_url="https://api.siliconflow.cn/v1"
            )
        retriever_from_llm = MultiQueryRetriever.from_llm(
            retriever=db.as_retriever(),
            llm=llm,
        )
        return retriever_from_llm.invoke(question)  #使用公共调用invoke()
        #print(retriever_from_llm)查看多角度扩展的配置及原理
        # results = retriever.invoke(question)

    def  chatWithDoc(self,question):
        _content=""
        context = self.askAndFindFiles(question)
        for i in context:
            _content += i.page_content
            
        messages = self.prompt.format_messages(context=_content,question=question)
        chat = ChatOpenAI(
            temperature=0,
            model="Pro/deepseek-ai/DeepSeek-V3.2",
            api_key="sk-oizcnnagplmevyplnagoxhggqqmpfprfhfoupaqvfvwonxes",
            base_url="https://api.siliconflow.cn/v1"
        )
        return chat.invoke(messages)

    

chat_doc = ChatDoc()
chat_doc.doc = "text1-docx.docx"
chat_doc.splitsentence()
docs = chat_doc.splitText
# print(chat_doc.splitText)

import requests
from typing import List, Optional
from langchain_core.embeddings import Embeddings #导入对接硅基流动API接口
from langchain_community.vectorstores import Chroma  #导入向量存储相关库

class SiliconFlowEmbeddings(Embeddings):
    """对接硅基流动(SiliconFlow) Embedding API的自定义类"""
    
    def __init__(
        self,
        api_key: str = "sk-oizcnnagplmevyplnagoxhggqqmpfprfhfoupaqvfvwonxes",
        model: str = "BAAI/bge-large-zh-v1.5",
        base_url: str = "https://api.siliconflow.cn/v1/embeddings",
        timeout: int = 30,
        max_retries: int = 2
    ):
        self.api_key = api_key
        self.model = model
        self.base_url = base_url
        self.timeout = timeout
        self.max_retries = max_retries
        
        # 创建一个带重试机制的Session
        self._session = requests.Session()
        adapter = requests.adapters.HTTPAdapter(max_retries=max_retries)
        self._session.mount('https://', adapter)
    
    def embed(self, texts: List[str]) -> List[List[float]]:
        """内部方法:调用SiliconFlow API获取向量"""
        headers = {
            "Authorization": f"Bearer {self.api_key}",
            "Content-Type": "application/json"
        }
        
        payload = {
            "model": self.model,
            "input": texts if len(texts) > 1 else texts[0],  # API支持单条或批量
            "encoding_format": "float"
        }
        
        try:
            response = self._session.post(
                self.base_url, 
                json=payload, 
                headers=headers,
                timeout=self.timeout
            )
            response.raise_for_status()
            result = response.json()
            
            # 提取向量数据
            embeddings = [item["embedding"] for item in result["data"]]
            return embeddings
            
        except requests.exceptions.RequestException as e:
            raise Exception(f"调用SiliconFlow API失败: {e}")
        
    def embed_documents(self, texts: List[str]) -> List[List[float]]:
        """嵌入文档列表 - 继承自Embeddings基类的抽象方法"""
        return self.embed(texts)
    
    def embed_query(self, text: str) -> List[float]:
        """嵌入查询文本 - 继承自Embeddings基类的抽象方法"""
        # 注意：这里text是单个字符串，但embed方法需要列表
        embeddings = self.embed([text])
        return embeddings[0]  # 返回单个向量的列表
    

#向量转化器设置
# embed(self, texts: List[str]) -> List[List[float]]:    
embeddings = SiliconFlowEmbeddings(
    api_key = "sk-oizcnnagplmevyplnagoxhggqqmpfprfhfoupaqvfvwonxes",
    model = "BAAI/bge-large-zh-v1.5",
    max_retries = 3
)
vectorstore = Chroma.from_documents(docs, embeddings)


#设置下logging查看生成查询
import logging
logging.basicConfig(level=logging.INFO)
logging.getLogger("langchain.retrievers.multi_query").setLevel(logging.DEBUG)
# unique_doc = chat_doc.askAndFindFiles("公司的名称是什么?")
# print(unique_doc)
# print(len(unique_doc))
chat_doc.chatWithDoc("这家公司的核心经营内容有什么？")



#print(vectorstore) #查看其在内存中的保存地址

# # 查看有多少个向量，确定向量化方式的正确性
# print(f"文档数量: {len(docs)}")
# print(f"向量存储中的向量数量: {vectorstore._collection.count()}")
# # 尝试一个简单的搜索
# results = vectorstore.similarity_search("它的预测容易出现稳定性差", k=3) #这里的k=3表示出现最接近的前三个检索结果
# print(f"搜索到 {len(results)} 个相关文档")
```
