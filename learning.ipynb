```python
import requests
from typing import List, Optional
from langchain_core.embeddings import Embeddings
class SiliconFlowEmbeddings(Embeddings):
    """对接硅基流动(SiliconFlow) Embedding API的自定义类"""
    
    def __init__(
        self,
        api_key: str,
        model: str = "BAAI/bge-large-zh-v1.5",
        base_url: str = "https://api.siliconflow.cn/v1/embeddings",
        timeout: int = 30,
        max_retries: int = 2
    ):
        self.api_key = "sk-oizcnnagplmevyplnagoxhggqqmpfprfhfoupaqvfvwonxes",
        self.model = model
        self.base_url = base_url
        self.timeout = timeout
        self.max_retries = max_retries
        
        # 创建一个带重试机制的Session
        self._session = requests.Session()
        adapter = requests.adapters.HTTPAdapter(max_retries=max_retries)
        self._session.mount('https://', adapter)
    
    def _embed(self, texts: List[str]) -> List[List[float]]:
        """内部方法:调用SiliconFlow API获取向量"""
        headers = {
            "Authorization": f"Bearer {self.api_key}",
            "Content-Type": "application/json"
        }
        
        payload = {
            "model": self.model,
            "input": texts if len(texts) > 1 else texts[0],  # API支持单条或批量
            "encoding_format": "float"
        }
        
        try:
            response = self._session.post(
                self.base_url, 
                json=payload, 
                headers=headers,
                timeout=self.timeout
            )
            response.raise_for_status()
            result = response.json()
            
            # 提取向量数据
            embeddings = [item["embedding"] for item in result["data"]]
            return embeddings
            
        except requests.exceptions.RequestException as e:
            raise Exception(f"调用SiliconFlow API失败: {e}")
    
    def embed_documents(self, texts: List[str]) -> List[List[float]]:
        """嵌入多个文档（用于构建知识库）"""
        return self._embed(texts)
    
    def embed_query(self, text: str) -> List[float]:
        """嵌入单个查询文本（用于检索）"""
        return self._embed([text])[0]

import os
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma
#from langchain_classic.chains import RetrievalQA

# 1. 初始化你的嵌入模型
embeddings = SiliconFlowEmbeddings(
    api_key="sk-oizcnnagplmevyplnagoxhggqqmpfprfhfoupaqvfvwonxes",  # 建议从环境变量读取
    model="BAAI/bge-large-zh-v1.5",  # 也可以换成 "BAAI/bge-m3" 等其他模型
    max_retries=3  # 增加重试次数，让程序更稳定
)

# 2. 加载并分割文档
loader = TextLoader("test.txt")
documents = loader.load()

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)
docs = text_splitter.split_documents(documents)

# 3. 创建向量数据库（这里会自动调用你的嵌入模型）
vectorstore = Chroma.from_documents(docs, embeddings)

# 4. 创建检索器
retriever = vectorstore.as_retriever()

# 5. 构建问答链（假设你也有一个对话模型）
# qa_chain = RetrievalQA.from_chain_type(
#     llm=your_llm,  # 可以是另一个自定义的SiliconFlow对话模型
#     retriever=retriever
# )
# 执行查询
# result = qa_chain.invoke("你的问题")
```
