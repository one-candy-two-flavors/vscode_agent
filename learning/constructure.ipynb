```python
#希尔排序
def insert_select_gap(data_set,gap):
    for i in range(gap,len(data_set)):
        tmp = data_set[i] #先将插入的数单独存起来，方便后面while循环中占用其位置
        j = i-gap
        while j>= 0 and data_set[j] > tmp:
            data_set[j+gap] = data_set[j]
            j -= gap
        data_set[j+gap] = tmp
def shell_sort(data_set):
    d = len(data_set) //2
    while d >= 1:
        insert_select_gap(data_set,d)
        d //= 2
data_set = list(range(1000))
import random
random.shuffle(data_set)
shell_sort(data_set)
print(data_set)
```

```python
#桶排序
def bucket_sort(data,n=100,max_num=10000):
    buckets = [[] for _ in range(n)] #创建桶，生成的列表元素为[]，所以这里是在创建一个二维的列表
    for val in data:
        i = min(val//(max_num//n),n-1)  #i表示放到几号桶里，（n-1用来限定范围，若为10000刚好超过最后的桶号）
        buckets[i].append(val)
        for j in range(len(buckets[i])-1,0,-1):
            if buckets[i][j] < buckets[i][j-1]:
                buckets[i][j],buckets[i][j-1] = buckets[i][j-1],buckets[i][j]
            else:
                break
    sorted_data = []
    for buc in  buckets:
        sorted_data.extend(buc)
    return sorted_data


import random
data = [random.randint(0,10000) for i in range(1000)]
data = bucket_sort(data)
print(data)

```

```python
#基数排序
def radix_sort(data):
    max_num = max(data)
    t = 0 
    while 10 ** t <= max_num:
        buckets = [[]for _ in range(10)]
        for val in data:
            buckets[val//(10 ** t)%10].append(val)
        data.clear()
        #把数重新写回data
        for buc in buckets:
            data.extend(buc)
        t += 1
import random
data = list(range(10000))
random.shuffle(data)
radix_sort(data)
print(data)
```

```python
class Queue:
    def __init__(self,size = 100):
        self.queue = [0 for _ in range(size)]
        self.size = size
        self.rear = 0 #队尾指针
        self.front = 0 #队首指针
    
    def push(self,element):
        if not self.is_filled():
            self.rear = (self.rear+1) % self.size
            self.queue[self.rear] = element
        else:
            raise IndexError("Queue is filled")

    
    def pop(self):
        if not self.is_empty():
            self.front = (self.front + 1) % self.size
            return self.queue[self.front]
        else:
            raise IndexError("Queue is empty.")
    
    #判断队空
    def is_empty(self):
        return self.rear == self.front
    
    #判断队满
    def is_filled(self):
        return (self.rear+1) % self.size == self.front
    
q = Queue(5)
for i in range(4):
    q.push(i)
print(q.is_filled)
```

```python
#2.背包问题
#①拿走的商品可以有分数情况
goods = [(60,10),(120,30),(100,20)]
goods.sort(key=lambda x:x[0]/x[1],reverse = True) 
#key=lambda x: x[0]/x[1]：这是一个匿名函数，对于每个元组 x，计算 x[0]除以x[1]的值。也就是说，计算第一个元素与第二个元素的比值。
#reverse=True：表示降序排序，即比值大的排在前面。

def fractional_backpack(goods,w):
    m = [0 for _ in range(len(goods))]
    for i,(prize,weight) in enumerate(goods):
        if w >=weight:
            m[i] = 1
            w -= weight
        else:
            m[i] = w/weight
            w = 0
            break
    return m

```

```python
activities = [(1,4),(3,5),(0,6),(5,7),(3,9)(5,9)(6,10)(8,11)(8,12)(2,14)(12,16)]
activities.sort(key = lambda x:x[1])#保证是按照结束时间最早结束的
def activity_selcection(a):
    res = [a[0]]#第一个活动
    for i in range(1,len(a)):
        if a[i][0] >= res[-1][1]:
            #不冲突
            res.append(a[i])
    return res
print(activity_selcection(activities))

```
